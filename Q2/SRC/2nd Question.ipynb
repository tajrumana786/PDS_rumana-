```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('2ndquesclean.csv')
```

```python
data
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['math score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='reading score', y='writing score', data=data, color='brown')
plt.title('Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='parental level of education', data=data, palette='pastel')
plt.title('Math Score vs English Skills')
plt.xlabel('English Skills')
plt.ylabel('Math Score')
plt.show()
```

```python
graph = sns.FacetGrid(data, col ='gender',hue_kws={'color': ['b', 'r']}, hue ='test preparation course') 
graph.map(plt.hist, 'race/ethnicity', bins = 15, color ='orange').add_legend()
```

```python
plt.figure(figsize=(8, 6)) 
correlation = data[['math score', 'reading score', 'writing score', 'english skills']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Scores')
plt.show()
```
