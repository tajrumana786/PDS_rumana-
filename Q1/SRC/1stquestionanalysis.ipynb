```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('clean_data_Fraility.csv')
```

```python
data
```

```python
sns.countplot(x='Age', data=data) 
plt.show()
```

```python
sns.boxplot(x='Frailty', y='Age', data=data) 
plt.title('Grip Strength by Frailty') 
plt.show()
```

```python
plt.plot(data['Age'], data['Grip strength'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.show()
```
